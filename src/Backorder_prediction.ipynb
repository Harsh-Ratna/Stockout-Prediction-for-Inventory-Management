### Handling Class Imbalance

```python
data = df_train['went_on_backorder'].value_counts()
data['No']
data['Yes']
```

```python
x = df_train.went_on_backorder.value_counts()
labels = ['no', 'yes']
print(x)
plt.pie(x, labels = labels, autopct='%.0f%%')
plt.title('Imbalanced classes')
```

Here, we can observe that only 1% of the cases went on backorder, and 99% did not. This is highly imbalanced and will greatly make the model biased. So now, we have to take a crucial decision, do we oversample or undersample data?
